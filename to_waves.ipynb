```python
import os
import subprocess



def process_audio_file(wav_path, output_dir, frame_rate=30):
    # 获取文件名（不含扩展名）和扩展名
    base_name = os.path.basename(wav_path)
    file_name, _ = os.path.splitext(base_name)
    # 打开日志文件
    out_file_path = os.path.join(output_dir, f"{file_name}.wav")
     # 使用 ffmpeg 將每個文件轉換為 WAV 格式
    subprocess.run(['ffmpeg', '-i', wav_path, "-acodec", "pcm_s16le", "-ac", "1", "-ar", "16000", out_file_path])

# 在 Jupyter Notebook 中指定输入和输出目录
input_dir = '/Users/jon/NTNU Dropbox/WiseSync/NAS/Datasets/ms_swift_taiwan_asr/RLHF/Audio/jin-jin-style'  # 替换为您的输入目录
output_dir = os.path.join(input_dir, 'wave')       # 替换为您的输出目录
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
frame_rate = 10  # 可以根据需要调整帧率

# 遍历输入目录中的所有 FLAC 文件
flac_files = [f for f in os.listdir(input_dir) if f.lower().endswith('.flac')]
flac_files.sort()
if not flac_files:
    print('在指定的输入目录中未找到任何 FLAC 文件。')
else:
    for flac_file in flac_files:
        wav_path = os.path.join(input_dir, flac_file)
        process_audio_file(wav_path, output_dir, frame_rate)
```
